````
predictions=model.predict(Xi_test)
````

````
T=Yi_test
````

````
TP,TN,FP,FN=0,0,0,0
for i in range(len(T)):
  if T[i]==1:
    if predictions[i]==1:
      TP+=1
    if predictions[i]==0:
      FN+=1
  if T[i]==0:
      if predictions[i]==1:
        FP+=1
      if predictions[i]==0:
        TN+=1
````

````
print(np.array([[TP,FP],[FN,TN]]))
````

````
precision=TP/(TP+FP)
print("precision=",precision)
````

````
recall=TP/(TP+FN)
print("recall=",recall)
````

````

F1_score=2*precision*recall/(precision+recall)
print("F1_score=",F1_score)
````
